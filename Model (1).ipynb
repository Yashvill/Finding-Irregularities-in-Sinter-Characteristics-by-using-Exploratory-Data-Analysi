```python
filtered=df[(df['Al2O3']>3.25)].index
df.drop(filtered,inplace=True)
df.shape
```

```python
df.head()
```

```python
from sklearn.preprocessing import MinMaxScaler
from sklearn.pipeline import Pipeline
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_absolute_error
from sklearn.model_selection import cross_val_score
```

```python
features=np.array(df.drop(['(CaO+MgO)/SiO2','T I(6.3 mm)','RDI'], axis=1))
labels=np.array(df["RDI"])
print(features.shape, labels.shape)
```

```python
xtrain,xtest,ytrain,ytest=train_test_split(features,labels,test_size=0.2)
print(xtrain.shape,ytrain.shape,xtest.shape,ytest.shape)
```

```python
Model=Pipeline([('scaling',MinMaxScaler()),('model',RandomForestRegressor(random_state=42))])
Model.fit(xtrain,ytrain)
```

```python
Model.score(xtrain,ytrain)
```

```python
pred=Model.predict(xtrain)
```

```python
predictions=Model.predict(xtest)
```

```python
mean_absolute_error(ytest, predictions)
```

```python
plt.scatter(ytrain,pred)
plt.plot(ytrain, ytrain, "r-")
plt.xlabel("Actual Value")
plt.ylabel("Predicted Value")
```

```python
plt.scatter(ytest,predictions)
plt.plot(ytest, ytest, "r-")
plt.xlabel("Actual Value")
plt.ylabel("Predicted Value")
```

```python
scores=-cross_val_score(Model,xtrain,ytrain,scoring="neg_mean_squared_error", cv=10)
print(scores)
print("mean:", scores.mean())
```
